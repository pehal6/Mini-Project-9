# Import required libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the data

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
data = pd.read_csv(url, sep=";")
```

# Show first 5 rows

```python
print("First 5 rows of the dataset:")
print(data.head())
```

# Basic info

```python
print("\nDataset shape (rows, columns):", data.shape)
print("\n Data types and non-null values:")
print(data.info())
```

# Check for missing values

```python
print("\n Missing values in each column:")
print(data.isnull().sum())
```


# Describe the data

```python
print("\n Statistical summary:")
print(data.describe())
```




#  Plot the count of different wine quality scores

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='quality', data=data)
plt.title("Wine Quality Distribution")
plt.xlabel("Quality Score")
plt.ylabel("Count")
plt.show()
```


# Correlation between columns

```python
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap="YlGnBu")
plt.title("Feature Correlation Heatmap")
plt.show()
```


# Alcohol vs. Wine Quality

```python
plt.figure(figsize=(8, 5))
sns.barplot(x='quality', y='alcohol', data=data)
plt.title("Average Alcohol Content by Wine Quality")
plt.xlabel("Wine Quality")
plt.ylabel("Average Alcohol")
plt.show()
```
